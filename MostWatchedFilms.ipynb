#Vamos para o Pandas!

```python
# lendo a base de uma página na web
df = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_filmes_de_maior_bilheteria',match='Distribuidor')[0]
df
```

```python
# Apenas os 3 primeiros mais assistidos
df[:3]
```

```python
#Os últimos 2 mais assistidos (top 100 filmes)
df[98:]
```

```python
#A lista em ordem alfabética
df.sort_values(by='Filme') # ordenando pela coluna 'Filme'
```

```python
#Em que posição está “O rei leão”
df.loc[(df['Filme'] == 'O Rei Leão')]
```
